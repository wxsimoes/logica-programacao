**Condicional**

```python
# if/else
num1 = 10
num2 = 5
if num1> num2:
  print( "O num1 é o maior número")

else:
    print("O num2 é o maior número")
```

```python
#if (se)/ elif(ou) / else(senão)

num1 = 50
num2 = 10
if num1 > num2:
  print( "O num1 é o maior número")
elif num1  < num2:
  print(" O num2 é o maior número")
else:
  print("Os dois são iguais")
```

COMANDO **FOR**


```python
# For Range
for contador in range(10):
  print(contador)
```

```python
# For range + Break (parar)
for contador in range(1000):
  print(contador)
  if contador ==2: #== (igual)
    break
```

```python
lista = ['Teste1', 'Teste2']
```

```python
#For simples
for val in lista:
  print(val)
```

```python
# For com Enumerate
for i, val  in enumerate(lista):
  print(i, val)
```

**Imagine que você está desenvolvendo um programa básico para verificar se um usuário pode acessar um sistema mediante digitação de senha.Veja como você pode resolver:**

```python
#Verificação de senha simples
#Definir a senha permitida
senha_permitida = '12345'
```

```python
#Solicitar a senha do usuário
senha_usuário = input('Digite a senha: ')
```

```python
#Verificar se a senha está correta
if senha_usuário == senha_permitida:
  print('Acesso permitido!')
else:
  print('Senha incorreta.Acesso negado.')
```
